### Data Cleaning 
Apabila pada proses sebelumnya kita menemukan masalah (missing value, outlier, dll.) yang terdapat di dalam sebuah data, masalah tersebut harus dibersihkan sebelum masuk tahap analisis data. Terdapat beberapa teknik yang dapat kita gunakan untuk membersihkan data. Seluruh teknik tersebut akan kita pelajari pada beberapa materi ke depan.

pada tahapan data cleaning (pembersihan data) ini dilakukan : 
- Penghapusan kolom yang tidak berhubungan dan tidak digunakan yaitu :
  - Kolom "Instan" (Karena tidak memiliki hubungan apapun)
  - Kolom "Workingday" (Tidak digunakan karena sudah di representasi kan oleh kolom weekday)
- Pengubahan tipe data integer ke category pada kolom :
  - season, month, weekday, holiday, weathersit (agar value didalam kolom tersebut dapat diubah)
- Pengubahan tipe data object ke datetime pada kolom dteday
- Pengubahan nama value dan nama kolom agar informasi dapat diterima lebih jelas :
  - Kolom dteday, mnth, hr, weathersit, windspeed, temp, atemp, hum, cnt
  - Value pada kolom season, year, month, weekday, holiday, weathersit
- Membuat kolom baru untuk mengkategorikan weekend dan weekday

```python
# menghapus kolom workingday
day_df.drop(['workingday'], axis = 1, inplace= True)
hour_df.drop(['workingday'], axis = 1, inplace= True)

# menghapus kolom instan
day_df.drop(['instant'], axis = 1, inplace= True)
hour_df.drop(['instant'], axis = 1, inplace= True)
```

```python
columns = ['season','mnth','weekday','holiday','weathersit']
# mengubah tipe data int ke category pada dataframe day_df
for column in columns:
    day_df[column] = day_df[column].astype('category')
# mengubah tipe data int ke category pada dataframe hour_df
for column in columns:
    hour_df[column] = hour_df[column].astype('category')
```

```python
# mengubah tipe data object ke datetime pada kedua dataframe 
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
```

```python
# mengubah nama kolom pada dataframe day_df
day_df.rename(columns=
              {'dteday':'date',
               'mnth':'month',
               'yr':'year', 
               'weathersit':'weather_situation', 
               'temp':'temperatur',
               'atemp':'feeling_temperature',
               'hum':'humidity',
               'windspeed':'wind_speed',
               'cnt':'total_renters'},
              inplace=True)

# mengubah nama kolom pada dataframe hour_df
hour_df.rename(columns=
              {'hr':'hour',
               'dteday':'date',
               'mnth':'month',
               'yr':'year', 
               'weathersit':'weather_situation', 
               'temp':'temperatur',
               'atemp':'feeling_temperature',
               'hum':'humidity',
               'windspeed':'wind_speed',
               'cnt':'total_renters'},
              inplace=True)
```

```python
# mengubah value menjadi lebih informatif dan lebih jelas 

# Ubah Value Kolom Season
day_df['season'] = day_df['season'].replace({
    1:'Spring',
    2:'Summer',
    3:'Fall',
    4:'Winter'})
hour_df['season'] = hour_df['season'].replace({
    1:'Spring',
    2:'Summer',
    3:'Fall',
    4:'Winter'})

# Ubah Value Kolom Year
day_df['year'] = day_df['year'].replace({
    0:'2011',
    1:'2012'})
hour_df['year'] = hour_df['year'].replace({
    0:'2011',
    1:'2012'})

# Ubah value Kolom month
day_df['month'] = day_df['month'].replace({
    1:'January', 7:'July',
    2:'February', 8:'August',
    3:'March', 9:'September',
    4:'April', 10:'October',
    5:'May', 11:'November',
    6:'June', 12:'December'})
hour_df['month'] = hour_df['month'].replace({
    1:'January', 7:'July',
    2:'February', 8:'August',
    3:'March', 9:'September',
    4:'April', 10:'October',
    5:'May', 11:'November',
    6:'June', 12:'December'})

#Ubah value Kolom Weekday
day_df['weekday'] = day_df['weekday'].replace({
    0:'Sunday', 4:'Thursday',
    1:'Monday', 5:'Friday',
    2:'Tuesday', 6:'Saturday',
    3:'Wednesday'})
hour_df['weekday'] = hour_df['weekday'].replace({
    0:'Sunday', 4:'Thursday',
    1:'Monday', 5:'Friday',
    2:'Tuesday', 6:'Saturday',
    3:'Wednesday'})

#Ubah value Kolom Weather_situation
day_df['weather_situation'] = day_df['weather_situation'].replace({
    1:'Clear',
    2:'Mist',
    3:'Light Snow/Light Rain',
    4:'Storm'})
hour_df['weather_situation'] = hour_df['weather_situation'].replace({
    1:'Clear',
    2:'Mist',
    3:'Light Snow/Light Rain',
    4:'Storm'})

#Ubah Value Kolom Holiday 
day_df['holiday'] = day_df['holiday'].replace({
    0:'Not Holiday',
    1:'Holiday'})
hour_df['holiday'] = hour_df['holiday'].replace({
    0:'Not Holiday',
    1:'Holiday'})
```

```python
# Membuat kolom baru 
day_df['day_type'] = day_df['weekday'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
hour_df['day_type'] = hour_df['weekday'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
```

1. Apakah ada perbedaan dalam jumlah penyewaan sepeda pada pengguna kasual (casual) dan pengguna terdaftar (registered) pada hari kerja, akhir pekan, atau hari libur?

Dari visualisasi barplot ini di dapatkan informasi sebagai berikut :

2. Apakah kondisi cuaca memiliki dampak yang signifikan terhadap jumlah penyewaan sepeda oleh pengguna kasual (casual) dan pengguna terdaftar (registered)


```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

summary_weather = day_df.groupby('weather_situation')[['registered','casual']].sum().reset_index()

sns.barplot(
    x="registered", 
    y="weather_situation", 
    data=summary_weather, 
    palette=colors,
    ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Penyewa Registered", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(
    x="casual", 
    y="weather_situation", 
    data=summary_weather, 
    palette=colors,
    ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Penyewa Casual", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Perbandingan Cuaca antara Penyewa Registered & Penyewa Casual", fontsize=20)
plt.show()
```

Dari visualisasi barplot diatas, informasi yang didapatkan yaitu :
- Penyewa registered paling banyak menyewa sepeda pada saat cuaca cerah (clear) sebanyak 1,811,606 dan paling rendah ketika cuaca sedang gerimis atau hujan sebanyak 33,974
- Penyewa casual paling banyak menyewa sepeda pada saat cuaca cerah (clear) sebanyak 446,346 dan paling rendah ketika cuaca sedang gerimis atau hujan sebanyak 3,895

3. Apakah tren jumlah pengguna meningkat atau menurun dari tahun ke tahun?


```python
# Mengelompokkan data berdasarkan tahun dan menjumlahkan penyewa registered dan casual
yearly_riders = day_df.groupby('year').agg({
    'registered': 'sum',
    'casual': 'sum'
}).reset_index()

plt.figure(figsize=(10, 6))

plt.plot(
    yearly_riders['year'], 
    yearly_riders['registered'], 
    marker='o', 
    label='Registered', 
    color='blue')

plt.plot(
    yearly_riders['year'], 
    yearly_riders['casual'], 
    marker='o', 
    label='Casual', 
    color='green')

plt.title('Jumlah Penyewa (Registered & Casual) dari Tahun 2011 sampai 2012', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Jumlah Penyewa', fontsize=12)
plt.legend()
plt.show()
```

Dari visualisasi lineplot ini didapatkan informasi bahwasanya jumlah penyewa sepeda baik dari penyewa registered dan penyewa casual mengalami kenaikan dari tahun 2011 sampai tahun 2012

4. Bagaimana musim mempengaruhi jumlah penyewa pada pengguna kasual (casual) dan pengguna terdaftar (registered)?


```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

summary_season = day_df.groupby('season')[['registered','casual','total_renters']].sum().reset_index()

sns.barplot(
    x="registered", 
    y="season", 
    data=summary_season, 
    palette=colors,
    ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Penyewa Registered", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(
    x="casual", 
    y="season", 
    data=summary_season, 
    palette=colors,
    ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Penyewa Casual", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Perbandingan Musim antara Penyewa Registered & Penyewa Casual", fontsize=20)
plt.show()
```

```python
sns.barplot(
  x='season',
  y='total_renters',
  data=summary_season,
  palette='husl')
plt.title("Jumlah Pengguna di 4 Musim")
plt.xlabel("Musim")
plt.ylabel("Jumlah Pengguna")
plt.show()
```

Dari visualisasi barplot ini didapatkan informasi perbandingan musim, yaitu :
- Perbandingan 4 musim terhadap penyewa registered paling sedikit terjadi ketika musim semi (Spring) sebanyak 410,726 dan tertinggi pada musim gugur (Fall) sebanyak 835,038.
- Perbandingan 4 musim terhadap penyewa casual paling sedikit terjadi ketika musim semi (Spring) sebanyak 60,622 dan tertinggi pada musing gugur (Fall) sebanyak 226,091.

5. Apakah terdapat tren peningkatan atau penurunan penggunaan sepeda setiap bulan?

```python
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Januari','February','March','April','May','June','July','August','September','October','November','December'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month","year"]).agg({
    "total_renters": "sum"
}).reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=monthly_counts,
    x="month",
    y="total_renters",
    hue="year",
    palette="rocket",
    marker="o")

plt.title("Jumlah total sepeda yang disewakan berdasarkan Bulan dan tahun")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

Dari visualisasi lineplot diatas didapatkan informasi sebagai berikut :
- Penyewaan tertinggi pada tahun 2011 terjadi pada bulan Juni sebanyak 143,512 penyewa dan terendah terjadi pada bulan januari sebanyak 0 penyewa
- Penyewaan tertinggi pada tahun 2012 terjadi pada bulan September sebanyak 218,573 dan terendah terjadi pada bulan januari sebanyak 0 penyewa
